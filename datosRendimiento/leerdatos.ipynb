```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lectura= pd.read_csv("datosfiltrados.csv")
```

```python
lectura.head()
```

```python
flagNA=lectura["Nota.Final"].isna()
flagNotNA=np.logical_not(flagNA)
lectura["Paso"]=False
lectura.loc[flagNotNA,"Paso"]=lectura["Nota.Final"][flagNotNA].str.split(',').str[-1].str.split('-').str[-1].astype('int')>=2
```

```python
np.sum(lectura.Paso.values)
```
