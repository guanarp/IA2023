```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parametros de las circunferencias
x0 = 4
y0 = 4
r0 = 2

x1 = 4
y1 = 4
r1 = 4

# Coordenadas polares aleatorias de ambas areas
ang = np.arange(0, 2*np.pi, 0.05)    # Puntos de una circunferencia en paso 0.05 rad

ptos_A1 = np.random.rand(ang.shape[0])*r0   # Puntos de la circunferencia 1
ptos_A2 = (np.random.rand(ang.shape[0])+1)*r1   # Puntos de la circunferencia 2

# Pasamos a coordenadas cartesianas los puntos aleatorios
x_int = ptos_A1*np.cos(ang) + x0
y_int = ptos_A1*np.sin(ang) + y0

x_ext = ptos_A2*np.cos(ang) + x1
y_ext = ptos_A2*np.sin(ang) + y1
```

```python
plt.plot(x_int, y_int, 'x')
plt.plot(x_ext, y_ext, 'o')
```

```python
z_int = np.ones((ang.shape[0],))
z_ext = np.zeros((ang.shape[0],))
```

receordando que la ecuacion de una cia es:

    theta4 * x4 + theta3 * x3 + theta2 * x2 + theta1 * x1 + theta0 * x0

```python
B = np.ones((ang.shape[0]*2,))
x = np.concatenate((x_int,x_ext),axis=0)
y = np.concatenate((y_int,y_ext),axis=0)
x_square = x**2
y_square = y**2
Z = np.concatenate((z_int,z_ext),axis=0)
print(Z.shape)
X = np.vstack((B,x,y,x_square,y_square)).T
X.shape
print(Z)
print(X)
```

```python
Y = Z.T
Y
#print(Y.shape)
```

```python
def sigmoide(z):
    return 1./(1+np.exp(-z)) #la z es de 252x1
```

```python
max_iter = 1000
m = (X.shape[0])
#print(m)
alpha = 0.01
theta = np.array(np.random.rand(5,))
J = []
for iter in range(max_iter):
    Z = np.dot(theta, np.transpose(X))
    #Z = np.dot(X, theta)
    #print(Z)
    h = sigmoide(Z)
    #print(h)
    #J = np.sum((Y - h)**2) / (2*m)
    #print(1/m * sum(-Y*np.log(h+0.00000001)) - sum((1-Y)*(np.log(1-h+0.00000001))))
    #J.append(1/m * np.dot(-Y,np.log(h+0.00000001)) - np.dot((1-Y),(np.log(1-h+0.00000001))))
    J.append(1/m * sum(-Y*np.log(h+0.00000001)) - sum((1-Y)*(np.log(1-h+0.00000001))))
    dj = -(h-y).dot(X)/m
    #print(dj)
    #print(np.sum(np.dot((y_data - h),x)))
    theta -= alpha * dj
```

```python
def pred(x,y):
    Z = x.dot(theta)
    h = sigmoide(Z)
    return h
```
